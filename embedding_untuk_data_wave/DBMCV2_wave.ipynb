```python
import pandas as pd
from sentence_transformers import SentenceTransformer
```

```python
# 0. Configuration
# -----------------------------------------------------------------------------
embedder = SentenceTransformer('sentence-transformers/distiluse-base-multilingual-cased-v2')
print("Model SentenceTransformer berhasil dimuat.")
```

```python
# 1. Load CSV file and convert 'menu' and 'embedding' to list
# -----------------------------------------------------------------------------
df = pd.read_csv('wave_emb_small.csv')
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
hasil = []
for idx, row in df.iterrows():
    alt = row['alt_resto'][2:-2]
    print(f"memproses {alt} di baris {idx+1}")
    emb = embedder.encode(alt)
    emb = [emb.tolist()]

    if emb:
        hasil.append({
            "restaurant": row["restaurant"],
            "menu": row["menu"],
            "city": row["city"],
            "alt_resto": row["alt_resto"],
            "embedding": emb
        })
        print("Embedding berhasil ditambahkan.")
    else:
        print("Embedding gagal ditambahakan.")

    print(f"Jumlah embedding saat ini: {len(hasil)}")

df_hasil = pd.DataFrame(hasil)
df_hasil
```
